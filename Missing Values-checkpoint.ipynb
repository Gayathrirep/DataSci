# Fill Null Values

```python
import pandas as pd
import numpy as np
import missingno as msno
```

```python
df = pd.read_csv('merged_data.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df.copy()
for col in df1:
    df1[col] = df1[col].fillna(df1[col].mode()[0])
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
msno.bar(df1)
```

```python
df1.to_csv('imputed_vals.csv', index=False)
```
